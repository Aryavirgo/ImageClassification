```python
import pandas as pd
import matplotlib.pyplot as plt
from skimage.transform import resize
from skimage.io import imread
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

```python
# img1=imread('/content/drive/MyDrive/Data/Cat/cat.17.jpg')
# img1.shape
```

```python
data_array=[]
target_array=[]
Categories=['Cat','Dog',]
datadir='/content/drive/MyDrive/Data'
for i in Categories:
  sub_path=os.path.join(datadir,i)
  images=os.listdir(sub_path)
  for img in images:
    img_path=os.path.join(sub_path,img)
    #print(img_path)
    img_array=imread(img_path)
    img_resized=resize(img_array,(150,150,3))
    data_array.append(img_resized.flatten())
    target_array.append(Categories.index(i))
  print("Loaded Category: ",i)
```

```python
import numpy as np
data=np.array(data_array)
data.shape
```

```python
target=np.array(target_array)
target.shape
```

```python
df=pd.DataFrame(data)
df
```

```python
df['target']=target
df
```

```python
df.isna().sum()
```

```python
df.dtypes
```

```python
X=df.iloc[:,:-1]
X
```

```python
y=df.iloc[:,-1]
y
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=.3,random_state=1)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier()
knn.fit(X_train,y_train)
y_pred=knn.predict(X_test)
print(y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

```python
sv=SVC()
sv.fit(X_train,y_train)
y_pr=sv.predict(X_test)
print(y_pr)
```

```python
print(classification_report(y_test,y_pr))
```
